```python
import cv2
import dlib
  
cap = cv2.VideoCapture('./video/Station.mp4') 
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))        # 取得畫面尺寸
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
# print(int(cap.get(cv2.CAP_PROP_POS_MSEC)))
fgbg = cv2.createBackgroundSubtractorMOG2()
fourcc = cv2.VideoWriter_fourcc(*'MGPG') #
out = cv2.VideoWriter('./video/hw_result.mp4', fourcc, 25.0, (width, height))
detector = dlib.get_frontal_face_detector() 
#原
while True:
    ret, frame = cap.read()
    if ret==True:    
        cv2.imshow('frame', frame)     
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 1500>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >= 1400:
            break  
        if k==27:
            break
    else:
        break   
# MOG2
while True:
    ret, frame = cap.read()
    frame = cv2.resize(frame, (width, height), interpolation = cv2.INTER_NEAREST)
    frame = fgbg.apply(frame)     
    cv2.imshow('frame', frame)
    out.write(frame)   
    k= cv2.waitKey(20) & 0xff
    if 3400>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >3350:
        break
    if k==27:
        break
#灰階
while True:
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    if ret==True:    
        cv2.imshow('frame', frame)     
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 7000>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >= 6950:
            break 
        if k == 27:
            break
    else:
        break  
#HSV
while True:
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    if ret==True:    
        cv2.imshow('frame', frame)     
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 9000>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) > 8950:
            break 
        if k == 27:
            break
    else:
        break
# RGB
while True:
    ret, frame = cap.read()
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    if ret==True:    
        cv2.imshow('frame', frame)     
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 11000>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) > 10950:
            break
        if k == 27:
            break
    else:
        break
#Canny 正
while True:
    ret, frame = cap.read()
    frame = cv2.Canny(frame, 100, 50)
    if ret==True:
        cv2.imshow('frame', frame)
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 15000>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >14950:
            break  
        if k == 27:
            break
    else:
        break
#Canny上下
while True:
    ret, frame = cap.read()
    frame = cv2.Canny(frame, 100, 50)
    if ret==True:    
        frame = cv2.flip(frame, -1)
        cv2.imshow('frame', frame)     
        out.write(frame)   
        k= cv2.waitKey(20) & 0xff
        if 17200>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) > 17150:
            break
        if k == 27:
            break
    else:
        break 
#Sobel
while True:
    ret, frame = cap.read()
    if ret==True: 
        x_grad = cv2.Sobel(frame,cv2.CV_32F, 1, 0)
        y_grad = cv2.Sobel(frame,cv2.CV_32F, 0, 1)              
        x_grad = cv2.convertScaleAbs(x_grad)
        y_grad = cv2.convertScaleAbs(y_grad)
        frame = cv2.add(x_grad, y_grad, dtype=cv2.CV_16S)
        frame = cv2.convertScaleAbs(frame)
        cv2.imshow('frame',frame)
        out.write(frame)
        k= cv2.waitKey(20) & 0xff
    if 25000>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >24950:
        break
    if k == 27:
        break
# 臉部
while(cap.isOpened()):                              # 以迴圈從影片檔案讀取影格，並顯示出來
    ret, frame = cap.read()
    if ret : # if true
        face_rects, scores, idx = detector.run(frame, 0, -0.5)  # 偵測人臉人臉分數大於-0.5才框起來
        for i, d in enumerate(face_rects):               # 取出所有偵測的結果
            x1 = d.left(); y1 = d.top(); x2 = d.right(); y2 = d.bottom()
            text = f'{scores[i]:.2f}, ({idx[i]:0.0f})'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 4, cv2.LINE_AA) # 以方框標示偵測的人臉
            cv2.putText(frame, text, (x1, y1), cv2.FONT_HERSHEY_DUPLEX,          # 標示分數
                        0.7, (255, 255, 255), 1, cv2.LINE_AA)
        out.write(frame)                               # 寫入影格
        cv2.imshow("frame", frame)           # 顯示結果
    k= cv2.waitKey(20) & 0xff
    if 30700>=int(cap.get(cv2.CAP_PROP_POS_MSEC)) >30650:
        break
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
